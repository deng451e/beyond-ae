```python
import matplotlib.pyplot as plt
import numpy as np

# Categories
labels = ['Beyond', 'Inifinigen', 'flexgen', 'H2O']
prefill_t = np.numpy([0.9656996950507164,0.6686409580288455,0.6583239609608427, 0.2170930930878967])
decode_t  = np.numpy([12.33962028787937,17.99327314819675,27.476321697351523,28.253102859365754])
 
dims = [2**n for n in range(1,5)]
bar_width = 0.35  # Width of the bars
index = np.arange(len(dims))  # X locations for the bars
 
fig,ax1 = plt.subplots(1, 1, figsize=(10, 4))

# --- Subfigure 1: Multiple bar with one stacked ---
# Plot the single bar data for subfigure 1
 
# Plot the stacked bar data for subfigure 1
 
ax1.bar(index + bar_width, prefill_t, bar_width, label='pcie transfer time', color='black')
ax1.bar(index + bar_width, prefill_t+decode_t, bar_width, bottom=prefill_t, label='gpu compute time', color='orange')




# Set title, labels, and legend for the first subplot
ax1.set_title('Latency,batch size 5, prefill 384, decode 128')
 
ax1.set_ylabel('Latency (sec)')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(dims)
ax1.legend()
 
# Adjust layout to prevent overlap
plt.tight_layout()

ax1.minorticks_off()
ax1.spines[['right', 'top']].set_visible(False)
 

 
# Display the figure
# plt.savefig('results-a6000/dtype-operation-time.pdf', bbox_inches='tight')

plt.show()
```
